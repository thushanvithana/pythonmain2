```python
X = data.drop('Outcome', axis=1)
y = data['Outcome']
```

```python
!pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train.shape, X_test.shape
```

```python
def model_acc(model):
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    print(str(model)+ ' -->' +str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model_acc(lr)

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
model_acc(dt)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Train Random Forest model
random_forest_model = RandomForestClassifier()
random_forest_model.fit(X_train, y_train)

# Evaluate model accuracy
accuracy = random_forest_model.score(X_test, y_test)
print("Random Forest Accuracy:", accuracy)
```

```python
from sklearn.svm import SVC

# Train SVM model
svm_model = SVC()
svm_model.fit(X_train, y_train)

# Evaluate model accuracy
accuracy = svm_model.score(X_test, y_test)
print("SVM Accuracy:", accuracy)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Train Gradient Boosting model
gradient_boosting_model = GradientBoostingClassifier()
gradient_boosting_model.fit(X_train, y_train)

# Evaluate model accuracy
accuracy = gradient_boosting_model.score(X_test, y_test)
print("Gradient Boosting Accuracy:", accuracy)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Train KNN model
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)

# Evaluate model accuracy
accuracy = knn_model.score(X_test, y_test)
print("KNN Accuracy:", accuracy)
```

```python
from sklearn.naive_bayes import GaussianNB

# Train Naive Bayes model
naive_bayes_model = GaussianNB()
naive_bayes_model.fit(X_train, y_train)

# Evaluate model accuracy
accuracy = naive_bayes_model.score(X_test, y_test)
print("Naive Bayes Accuracy:", accuracy)
```

```python
from sklearn.neural_network import MLPClassifier

# Train Neural Network model
neural_network_model = MLPClassifier()
neural_network_model.fit(X_train, y_train)

# Evaluate model accuracy
accuracy = neural_network_model.score(X_test, y_test)
print("Neural Network Accuracy:", accuracy)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# Define the hyperparameter grid
parameters = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
}

# Initialize GridSearchCV
grid_obj = GridSearchCV(estimator=lr, param_grid=parameters, cv=5)

# Fit GridSearchCV to find the best model
grid_fit = grid_obj.fit(X_train, y_train)

# Get the best model
best_model = grid_fit.best_estimator_
best_model
```

```python
import pickle
with open('predictor.pickle', 'wb') as file:
    pickle.dump(best_model, file)
```
